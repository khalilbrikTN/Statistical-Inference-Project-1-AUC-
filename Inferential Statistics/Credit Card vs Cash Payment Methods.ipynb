# Statistical Analysis: Credit Card vs Cash Payment Methods

**Research Question:** Do customers who pay with Credit cards have significantly different average transaction values compared to those who pay with Cash?

**Analyst:** Khalil

---

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load data
df = pd.read_csv('supermarket_sales.csv')
print(f"Dataset loaded: {df.shape[0]} transactions\n")
```

## 1. Descriptive Statistics

```python
# Separate data by payment method (Credit card vs Cash only)
credit_transactions = df[df['Payment'] == 'Credit card']['Total']
cash_transactions = df[df['Payment'] == 'Cash']['Total']

# Calculate statistics
credit_mean = credit_transactions.mean()
credit_std = credit_transactions.std()
credit_n = len(credit_transactions)

cash_mean = cash_transactions.mean()
cash_std = cash_transactions.std()
cash_n = len(cash_transactions)

credit_total = credit_transactions.sum()
cash_total = cash_transactions.sum()
total_sales = credit_total + cash_total

print("="*70)
print("DESCRIPTIVE STATISTICS")
print("="*70)
print(f"\n{'Payment':<15} {'Mean ($)':<15} {'Std Dev ($)':<15} {'Sample Size':<15}")
print("-"*70)
print(f"{'Credit card':<15} {credit_mean:<15.2f} {credit_std:<15.2f} {credit_n:<15}")
print(f"{'Cash':<15} {cash_mean:<15.2f} {cash_std:<15.2f} {cash_n:<15}")

print(f"\n{'='*70}")
print("TOTAL SALES")
print("="*70)
print(f"Credit card: ${credit_total:,.2f} ({credit_total/total_sales*100:.2f}%)")
print(f"Cash:        ${cash_total:,.2f} ({cash_total/total_sales*100:.2f}%)")
print(f"Difference: {abs(credit_total/total_sales - cash_total/total_sales)*100:.2f} percentage points")
```

## 2. Calculate Test Statistics

```python
# Step 1: Difference in sample means
mean_diff = credit_mean - cash_mean
print("="*70)
print("STEP 1: DIFFERENCE IN SAMPLE MEANS")
print("="*70)
print(f"x̄₁ - x̄₂ = {credit_mean:.2f} - {cash_mean:.2f} = {mean_diff:.2f}")

# Step 2: Pooled standard deviation
pooled_var = ((credit_n - 1) * credit_std**2 + (cash_n - 1) * cash_std**2) / (credit_n + cash_n - 2)
pooled_std = np.sqrt(pooled_var)
print(f"\n{'='*70}")
print("STEP 2: POOLED STANDARD DEVIATION")
print("="*70)
print(f"sp = √[((n₁-1)s₁² + (n₂-1)s₂²) / (n₁+n₂-2)]")
print(f"sp = √[(({credit_n}-1)×{credit_std:.2f}² + ({cash_n}-1)×{cash_std:.2f}²) / {credit_n+cash_n-2}]")
print(f"sp = {pooled_std:.2f}")

# Step 3: Standard error
se = pooled_std * np.sqrt(1/credit_n + 1/cash_n)
print(f"\n{'='*70}")
print("STEP 3: STANDARD ERROR")
print("="*70)
print(f"SE = sp × √(1/n₁ + 1/n₂)")
print(f"SE = {pooled_std:.2f} × √(1/{credit_n} + 1/{cash_n})")
print(f"SE = {pooled_std:.2f} × {np.sqrt(1/credit_n + 1/cash_n):.4f}")
print(f"SE = {se:.2f}")

# Step 4: 95% Confidence Interval
t_critical = 1.96  # For large samples
ci_lower = mean_diff - t_critical * se
ci_upper = mean_diff + t_critical * se
print(f"\n{'='*70}")
print("STEP 4: 95% CONFIDENCE INTERVAL")
print("="*70)
print(f"CI = (x̄₁ - x̄₂) ± t₀.₀₂₅,df × SE")
print(f"CI = {mean_diff:.2f} ± {t_critical} × {se:.2f}")
print(f"CI = {mean_diff:.2f} ± {t_critical * se:.2f}")
print(f"CI = [{ci_lower:.2f}, {ci_upper:.2f}]")
print(f"\nInterpretation: CI {'includes' if ci_lower <= 0 <= ci_upper else 'does not include'} 0")
if ci_lower <= 0 <= ci_upper:
    print("→ Suggests no significant difference")
else:
    print("→ Suggests significant difference exists")
```

## 3. Hypothesis Testing

```python
print("="*70)
print("HYPOTHESIS TEST")
print("="*70)
print("\nH₀: μ_Credit = μ_Cash (No difference in average transaction value)")
print("H₁: μ_Credit ≠ μ_Cash (Significant difference exists)")
print("\nSignificance level: α = 0.05")

# Calculate t-statistic
t_stat = mean_diff / se
df = credit_n + cash_n - 2

print(f"\n{'='*70}")
print("T-STATISTIC")
print("="*70)
print(f"t = (x̄₁ - x̄₂) / SE = {mean_diff:.2f} / {se:.2f} = {t_stat:.3f}")
print(f"df = n₁ + n₂ - 2 = {credit_n} + {cash_n} - 2 = {df}")

# Calculate p-value using scipy
t_test_result = stats.ttest_ind(credit_transactions, cash_transactions)
p_value = t_test_result.pvalue

print(f"\n{'='*70}")
print("P-VALUE")
print("="*70)
print(f"For two-tailed test with t = {t_stat:.3f} and df = {df}:")
print(f"p-value = {p_value:.4f}")
```

## 4. Effect Size

```python
# Cohen's d
cohens_d = mean_diff / pooled_std

print("="*70)
print("EFFECT SIZE (Cohen's d)")
print("="*70)
print(f"d = (x̄₁ - x̄₂) / sp = {mean_diff:.2f} / {pooled_std:.2f} = {cohens_d:.3f}")

if abs(cohens_d) < 0.2:
    effect = "Negligible"
elif abs(cohens_d) < 0.5:
    effect = "Small"
elif abs(cohens_d) < 0.8:
    effect = "Medium"
else:
    effect = "Large"

print(f"\nInterpretation: {effect} effect (d {'<' if abs(cohens_d) < 0.2 else '>='} 0.2)")
```

## 5. Decision and Conclusion

```python
print("="*70)
print("DECISION")
print("="*70)
print(f"\nDecision rule: Reject H₀ if p < 0.05")
print(f"\nSince p = {p_value:.4f} {'<' if p_value < 0.05 else '>'} 0.05:")

if p_value < 0.05:
    decision = "REJECT H₀"
    conclusion = f"There IS a statistically significant difference between credit card and cash transaction values."
else:
    decision = "FAIL TO REJECT H₀"
    conclusion = f"There is NO statistically significant difference between credit card and cash transaction values."

print(f"\n→ {decision}")

print(f"\n{'='*70}")
print("CONCLUSION")
print("="*70)
print(f"\n{conclusion}")
print(f"\nThe observed ${abs(mean_diff):.2f} difference {'is' if p_value < 0.05 else 'is not'} statistically significant.")

if p_value >= 0.05:
    print(f"This difference is likely attributable to random variation.")
else:
    print(f"Credit card customers spend ${abs(mean_diff):.2f} {'more' if mean_diff > 0 else 'less'} per transaction on average.")

print(f"\nEffect size: {effect} (Cohen's d = {cohens_d:.3f})")
```
